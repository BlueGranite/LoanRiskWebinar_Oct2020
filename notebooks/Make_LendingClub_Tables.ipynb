# LendingClub Data

https://www.kaggle.com/wordsforthewise/lending-club

```
# File location and type
file_location = "/FileStore/tables/lendingclub/accepted_2007_to_2018Q4_csv.gz"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
accepted = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(accepted)
```

```
permanent_table_name = "lendingclub_accepted"

accepted.write.format("parquet").saveAsTable(permanent_table_name)
```

```
from pyspark.sql.functions import col

# File location and type
file_location = "/FileStore/tables/lendingclub/rejected_2007_to_2018Q4_csv.gz"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
rejected = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

column_list = rejected.columns
new_column_list = ["amount_requested", "application_date", "load_title", "risk_score", "debt_to_income_ratio", "zip_code", "state", "employment_length", "policy_code"]
column_mapping = [[o, n] for o, n in zip(column_list, new_column_list)]

rejected = rejected.select(list(map(lambda old, new: col(old).alias(new),*zip(*column_mapping))))

display(rejected)
```

```
permanent_table_name = "lendingclub_rejected"

rejected.write.format("parquet").saveAsTable(permanent_table_name)
```
